```python
## Q&A,chatbots,text summarization
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("running", pos='v'))  # Output: run
```

```python
'''
POS(Parts of Speech)- Noun-n
verb-v
adjective-a
adverb-r
'''
```

```python
words = ["running", "better", "flies", "studies", "children", "mice", "geese", "happier", "swimming", "bought"]
print(words)
```

```python
for word in words:
    lemma_noun = lemmatizer.lemmatize(word, pos='n')
    lemma_verb = lemmatizer.lemmatize(word, pos='v')
    lemma_adj = lemmatizer.lemmatize(word, pos='a')
    lemma_adv = lemmatizer.lemmatize(word, pos='r')
    print(f"{word:12} | Noun: {lemma_noun:10} | Verb: {lemma_verb:10} | Adj: {lemma_adj:10} | Adv: {lemma_adv:10}")
```
