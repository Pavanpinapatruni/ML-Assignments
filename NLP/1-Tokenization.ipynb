## Tokenize
### Sentence from ParaGraphs

```python
from nltk.tokenize import sent_tokenize
```

the punkt_tab data to be downloaded for sentence tokenization

NLTK (Natural Language Toolkit) is a modular library - it doesn't install all language data by default to keep the initial installation lightweight. Instead, it downloads specific language resources only when you need them.

The sent_tokenize() function uses the Punkt tokenizer algorithm to split text into sentences. This algorithm needs:

1. <b>Pre-trained models</b> that understand sentence boundaries for specific languages.
2. <b>Statistical data</b> about punctuation patterns, abbreviations, and sentence endings.
3. <b>Language-specific rules</b> for different languages (English, Spanish, etc.)
The punkt_tab resource contains:

- Trained models for detecting sentence boundaries
- Abbreviation lists (like "Dr.", "Mr.", "U.S.A.")
- Statistical patterns for punctuation usage
- Rules for handling edge cases in sentence segmentation
  

<b>Without downloading this data, NLTK can't distinguish between:</b>  

- Period ending a sentence vs. period in an abbreviation ("Dr. Smith")
- Ellipsis vs. sentence ending ("Wait...")
- Decimal numbers vs. sentence boundaries ("Price is $3.50. Next item...")


```python
import nltk
nltk.download('punkt_tab')
```

```python
documents=sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```
